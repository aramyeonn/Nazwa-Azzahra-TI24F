```python
#MATplotlib, SeaBorn
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
np.random.seed(42)
bulan = ['Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni', 'Juli', 'Agustus', 'September', 'Oktober', 'November', 'Desember']
Min ggu_1 = np.random.randint(500000, 700001, 12)
Minggu_2 = np.random.randint(500000, 700001, 12)
Minggu_3 = np.random.randint(500000, 700001, 12)
Minggu_4 = np.random.randint(500000, 700001, 12)

data ={
    'Bulan' : bulan,
    'Minggu1' : Minggu_1,
    'Minggu2' : Minggu_2,
    'Minggu3' : Minggu_3,
    'Minggu4' : Minggu_4
}
df = pd.DataFrame(data)
df['Total_Pengeluaran'] = df[['Minggu1','Minggu2','Minggu3','Minggu4']].sum(axis=1)
df
```

```python
#Diagram Bar
#Buat canvas
plt.figure(figsize=(12,6)) #widht, height -> inch
plt.bar(df['Bulan'],df['Total_Pengeluaran'],color='pink')
plt.xticks(rotation=45)
plt.tight_layout()
plt.title("Total Pengeluaran Mahasiswa Per Bulan")
plt.ylabel("Rupiah")
plt.show()
```

```python
#Tren pengeluaran Per Minggu dalam satu tahun

plt.figure(figsize=(10,10))

for minggu in ['Minggu1','Minggu2','Minggu3','Minggu4']:
    plt.plot(df['Bulan'],df[minggu],marker='o',label=minggu)

plt.title("Tren Pengeluaran Mahasiswa per Minggu")
plt.ylabel("Rupiah")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Diagram Pie ( Proposi Pengeluaran Per Bulan )
plt.figure(figsize=(6,6))
plt.pie(df['Total_Pengeluaran'],labels=df['Bulan'], startangle=90, autopct='%1.1f%%',colors=plt.cm.Pastel1.colors)
plt.axis('equal')
plt.title("Proporsi Pengeluaran Mahasiswa Per Bulan")
plt.show()      
```

```python
#Model Ai -> scikit , sklearn
from sklearn.linear_model import LinearRegression
df['bulan'] = np.arange(1,13)

x = df[['bulan']]
y = df['Total_Pengeluaran']

model = LinearRegression()
model.fit(x,y)

prediksi_13 = model.predict([[13]])[0]
print(f"Prediksi Pengeluaran Mahasiswa bulan Januari: {prediksi_13:,.0f}")
plt.figure(figsize=(10,10))
plt.boxplot(df['Total_Pengeluaran'])
plt.show()
```
